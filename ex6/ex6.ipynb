```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Directed Graph

```python
# Create an empty directed graph
G = nx.DiGraph()
```

```python
# Add edges to the graph
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 1)
```

```python
# Draw the graph
nx.draw(G, with_labels=True)
```
