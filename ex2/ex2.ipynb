```python
import numpy  as np
import networkx as nx
import inspect
```

```python
# Generate a regular graph using the network module
G = nx.random_regular_graph(3, 10)
```

```python
G.nodes()
```

```python
# visualize the graph
nx.draw(G)
```

```python
#Generate a laplacian matrix
L = nx.laplacian_matrix(G)
```

```python
L.toarray()
```

```python
A = nx.adjacency_matrix(G).toarray()
D = np.diag(np.ones(10)*3)
L1 = D - A
```

```python
L1
```

```python
np.array_equal(L.toarray(), L1)
```

```python
#Ex2.2
# From given Laplace matrix, find the D and A matrix
```

```python
L = np.array([[ 2, -1, -1, 0],[-1, 3, -1, -1],[-1, -1, 3, -1],[0, -1, -1, 2]])
L
```

```python
# find the D matrix
D = np.diag(np.diag(L))
```

```python
D
```

```python
A = D - L
```

```python
A 
```

```python
# plot the graph
G = nx.from_numpy_matrix(A)
nx.draw(G)
```
